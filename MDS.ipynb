```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('/mnt/data/mall-customers-data.csv')

# Selecting features for clustering
X = data[['annual_income', 'spending_score']]

# Calculate the Within-Cluster-Sum-of-Squares (WCSS) for different numbers of clusters
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

# Plot the Elbow graph
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('mall-customers-data.csv')

# Selecting features for clustering
X = data[['annual_income', 'spending_score']]

# Perform KMeans clustering with 3 clusters
kmeans = KMeans(n_clusters=3, init='k-means++', random_state=42)
data['Cluster'] = kmeans.fit_predict(X)

# Plotting the clusters
plt.scatter(X['annual_income'], X['spending_score'], c=data['Cluster'], cmap='viridis', marker='o')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', marker='X', label='Centroids')
plt.title('Customer Segments with K=3')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
```

```python
from sklearn.cluster import DBSCAN

# Applying DBSCAN with specified parameters
dbscan = DBSCAN(eps=5, min_samples=5)
data['Cluster'] = dbscan.fit_predict(X)

# Plotting the DBSCAN clusters
plt.figure(figsize=(8, 6))
plt.scatter(X['annual_income'], X['spending_score'], c=data['Cluster'], cmap='viridis', marker='o')
plt.title('Customer Segments using DBSCAN')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.grid(True)
plt.show()

```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt

# Read the data
df = pd.read_csv('mall-customers-data.csv')

# Select features for clustering
features = ['annual_income', 'spending_score']
X = df[features].values

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply DBSCAN
# These parameters can be tuned based on the desired clustering outcome
dbscan = DBSCAN(eps=0.3, min_samples=5)
clusters = dbscan.fit_predict(X_scaled)

# Add cluster labels to the dataframe
df['Cluster'] = clusters

# Create visualization
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df['annual_income'], df['spending_score'],
                     c=df['Cluster'], cmap='viridis')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('DBSCAN Clustering of Mall Customers')
plt.colorbar(scatter, label='Cluster')

# Print summary statistics
print("\nNumber of clusters found:", len(set(clusters)) - (1 if -1 in clusters else 0))
print("Number of noise points:", list(clusters).count(-1))

# Print cluster statistics
print("\nCluster Statistics:")
for cluster in sorted(set(clusters)):
    cluster_size = len(df[df['Cluster'] == cluster])
    avg_income = df[df['Cluster'] == cluster]['annual_income'].mean()
    avg_spending = df[df['Cluster'] == cluster]['spending_score'].mean()
   
    if cluster == -1:
        print(f"\nNoise points:")
    else:
        print(f"\nCluster {cluster}:")
    print(f"Size: {cluster_size}")
    print(f"Average Income: ${avg_income:.2f}k")
    print(f"Average Spending Score: {avg_spending:.2f}")

plt.show()
```
